# Distributions

## Story 1 -- Heights of Students
The heights of students in a high school are known to follow a normal distribution with a mean height of 170 cm and a standard deviation of 10 cm. Write Python code to generate a sample of 1000 student heights following this distribution, calculate the mean and variance of the generated sample, and interpret the results.

```python
import numpy as np

mean = 170
std_dev = 10
num_samples = 1000

np.random.seed(24)

samples = np.random.normal(mean, std_dev, num_samples)

mean = np.mean(samples)
variance = np.var(samples)
std = np.std(samples)

print(f'\n Mean: {mean}' +
      f'\n Variance: {variance}' +
      f'\n Standard Deviation: {std}')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(7, 5))
sns.histplot(samples, bins=50, kde=True, color='skyblue')
plt.title('Height Distribution of Students')
plt.xlabel('Height (cm)')
plt.ylabel('Probability Density')
plt.show()
```

### *Conclusion*
The generated data of mean $170.136$ and standard deviation $10.132$ is very close to the theoretical input of $170$ and $10$. The kernel density estimation (KDE) line shows a fairly clean normal distribution of student heights centered around a mean of $170.136 cm$.

## Story 3 -- Call Center Inbound Calls
A call center receives an average of 3 inbound calls per 10 minutes. Write Python code to generate 500 samples from this Poisson distribution, calculate the mean and variance, and iterpret their significance for the call center’s operations.

```python
from scipy.stats import poisson

np.random.seed(24)

lambda_value = 3
poisson_samples = np.random.poisson(lam=lambda_value, size=500)

mean = np.mean(poisson_samples)
variance = np.var(poisson_samples)

print(f'\n Mean: {mean}' +
      f'\n Variance: {variance}')
```

```python
plt.hist(poisson_samples, bins=10, edgecolor='black', alpha=0.7, density=True, label='Sample Data')

x = np.arange(0, max(poisson_samples) + 1)
pmf_values = poisson.pmf(x, mu=lambda_value)

plt.plot(x, pmf_values, marker='o', color='red', label='Theoretical PMF')

plt.title('Distribution of Call Center Inbound Calls')
plt.xlabel('Number of Calls per 10 Minutes')
plt.ylabel('Density')
plt.legend()
plt.show()
```

## Story 4 -- Typos in a Book Chapter
The editor of a textbook finds that on average, 2 typos appear per 50 pages. Write Python code to simulate 2000 samples from the corresponding Poisson distribution, calculate the mean and variance of the samples, and interpret the findings for the editor’s review process.

```python
lambda_value = 2
poisson_samples = np.random.poisson(lam=lambda_value, size=2000)

mean = np.mean(poisson_samples)
variance = np.var(poisson_samples)

print(f'\n Mean: {mean}' +
      f'\n Variance: {variance}')
```

```python
plt.hist(poisson_samples, bins=9, edgecolor='black', alpha=0.7, density=True, label='Sample Data')

x = np.arange(0, max(poisson_samples) + 1)
pmf_values = poisson.pmf(x, mu=lambda_value)

plt.plot(x, pmf_values, marker='o', color='red', label='Theoretical PMF')

plt.title('Distribution of Typos in a Book')
plt.xlabel('Typos per 50 Pages')
plt.ylabel('Density')
plt.legend()
plt.show()
```
