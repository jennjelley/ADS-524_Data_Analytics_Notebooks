## Story 1 -- Evaluating the Effectiveness of Three Study Techniques
Suppose a researcher wants to test the effect of three different study techniques on student performance
in a math exam: group study, self-study, and tutoring. Assume that the students were randomly assigned to
one of the study techniques and that the groups are equivalent in terms of prior math knowledge and study
habits. The researcher wants to determine if there is a significant difference in the exam scores between
the three study techniques. Since there are three groups, the researcher uses One-Way ANOVA to test if
the mean exam scores differ significantly between the three study techniques.
- Group 1 - Group Study: [85, 90, 92, 88, 91, 85, 90, 92, 88, 91]
- Group 2 - Self Study: [75, 78, 77, 74, 80, 75, 78, 77, 74, 80]
- Group 3 - Tutoring: [95, 98, 97, 94, 96, 95, 98, 97, 94, 96]

---

**Hypotheses**
- $H_0$: The **Null Hypothesis** is that none of the three study techniques have an impact on test scores.
- $H_1$: The **Alternative Hypothesis** is that study technique has an impact on test scores.

**F-statistic and p-value**

```python
from scipy import stats

group_group = [85, 90, 92, 88, 91, 85, 90, 92, 88, 91]
group_self = [75, 78, 77, 74, 80, 75, 78, 77, 74, 80]
group_tutor = [95, 98, 97, 94, 96, 95, 98, 97, 94, 96]

f_statistic, p_value = stats.f_oneway(group_group, group_self, group_tutor)

print("F-statistic: ", f_statistic, "\np-value: ", p_value)
```

### Interpret

A large F-statistic, $201.1698$, tells us that there is a much greater deal of variability between students in different groups than there is between students within the same group. A p-value less than $0.05$, $6.0339e{-17}$ allows us the to say this with a 95% confidence level.

**Visualize**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(9, 3))

sns.kdeplot(group_self, color='green', label='Self Study', fill=True)
sns.kdeplot(group_group, color='blue', label='Group Study', fill=True)
sns.kdeplot(group_tutor, color='red', label='Tutoring', fill=True)

plt.yticks([])
plt.xlabel('Score')
plt.title('KDE Plot of Study Techniques')
plt.legend()

plt.show()
```

## Story 2 -- Evaluating the Effectiveness of Diet Programs
Suppose a researcher wants to test the effect of three different diets on weight loss: Diet A, Diet B, and Diet C. The researcher randomly assigns 15 participants to each diet group, ensuring that the groups are equivalent in terms of age and initial weight. After one month, the researcher measures the amount of weight lost by each participant. The data is as follows:
- Diet A: [3, 4, 5, 2, 3, 4, 6, 7, 5, 4, 3, 6, 5, 4, 3]
- Diet B: [1, 2, 2, 3, 1, 1, 2, 2, 3, 3, 1, 2, 1, 2, 2]
- Diet C: [6, 7, 8, 7, 6, 8, 9, 8, 7, 9, 8, 7, 6, 7, 8]

The researcher wants to determine if there is a significant difference in weight loss between the three diets. Since there are three diet groups, the researcher uses a One-Way ANOVA to test if the mean weight loss differs significantly between the diets.  

---

**Hypotheses**
- $H_0$: The **Null Hypothesis** is that none of the three diets have an impact on weight loss.
- $H_1$: The **Alternative Hypothesis** is that at least one diet has an impact on weight loss.

**F-statistic and p-value**

```python
from scipy import stats

group_A = [3, 4, 5, 2, 3, 4, 6, 7, 5, 4, 3, 6, 5, 4, 3]
group_B = [1, 2, 2, 3, 1, 1, 2, 2, 3, 3, 1, 2, 1, 2, 2]
group_C = [6, 7, 8, 7, 6, 8, 9, 8, 7, 9, 8, 7, 6, 7, 8]

f_statistic, p_value = stats.f_oneway(group_A, group_B, group_C)

print("F-statistic: ", f_statistic, "\np-value: ", p_value)
```

### Interpret

An F-statistic, $100.4945$, tells us that there is some variability and some cross-over in weight loss between participants in the three groups. A p-value less than $0.05$, $9.7931e{-17}$ allows us the to say this with a 95% confidence level.

**Visualize**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(9, 3))

sns.kdeplot(group_A, color='green', label='Group A', fill=True)
sns.kdeplot(group_B, color='blue', label='Group B', fill=True)
sns.kdeplot(group_C, color='red', label='Group C', fill=True)

plt.yticks([])
plt.xlabel('Weight Loss')
plt.title('KDE Plot of Diet Programs')
plt.legend()

plt.show()
```

## Story 3 -- Evaluating the Two Factors on Employee Productivity
A researcher is studying the effects of two factors on employee productivity: Training Type and Work
Hours. The two training types are Online Training (OT) and In-Person Training (IP), and the two work
hour levels are 6 hours and 8 hours. The researcher collects productivity scores for 10 employees from
each combination of factors, resulting in the following data:
- Online Training, 6 hours: 50, 55, 52, 58, 54, 60, 53, 59, 61, 56
- Online Training, 8 hours: 70, 74, 75, 72, 78, 80, 76, 77, 79, 81
- In-Person Training, 6 hours: 45, 50, 48, 53, 49, 51, 50, 52, 47, 55
- In-Person Training, 8 hours: 65, 70, 72, 68, 74, 77, 73, 75, 78, 79

**Hypotheses**
- $H_0$: The **Null Hypothesis** is there is no significant effect of training type or work hours on employee productivity.
  - $H_0^{Training}$ There is no effect of training type on employee productivity.
  - $H_0^{Hours}$ There is no effect of work hours on employee productivity.
  - $H_0^{Interaction}$ There is no interaction effect between training type and work hours.
- $H_1$: The **Alternative Hypothesis** is at least one factor or their interaction has a significant effect on employee productivity.
  - $H_1^{Training}$ There is a significant effect of training type on employee productivity.
  - $H_1^{Hours}$ There is a significant effect of work hours on employee productivity.
  - $H_1^{Interaction}$ There is a significant interaction effect between training type and work hours.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

group_IP_6 = pd.DataFrame({'Training Type':'In-Person Training','Work Hours':'6','Productivity Score':[45, 50, 48, 53, 49, 51, 50, 52, 47, 55]})
group_IP_8 = pd.DataFrame({'Training Type':'In-Person Training','Work Hours':'8','Productivity Score':[65, 70, 72, 68, 74, 77, 73, 75, 78, 79]})
group_OT_6 = pd.DataFrame({'Training Type':'Online Training','Work Hours':'6','Productivity Score':[50, 55, 52, 58, 54, 60, 53, 59, 61, 56]})
group_OT_8 = pd.DataFrame({'Training Type':'Online Training','Work Hours':'8','Productivity Score':[70, 74, 75, 72, 78, 80, 76, 77, 79, 81]})
df = pd.concat([group_IP_6, group_IP_8, group_OT_6, group_OT_8])


plt.figure(figsize=(12, 5))
sns.boxplot(x='Training Type',y='Productivity Score',hue='Work Hours',data=df)
plt.title('Productivity Score by Training Type and Work Hours')
plt.show()
```

```python
df = df.rename(columns=lambda s: s.replace("_", " "))

sns.pointplot(x='Training Type', y='Productivity Score', hue='Work Hours', data=df, errorbar=('ci', False))
plt.title('Interaction of Training Type and Work Hours on Productivity')
plt.show()
```

```python
import statsmodels.api as sm

df = df.rename(columns=lambda s: s.replace(" ", "_"))

model = ols('Productivity_Score ~ C(Training_Type) + C(Work_Hours) + C(Training_Type):C(Work_Hours)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
anova_table
```

**Interpretation**  
There is a significant main effect of training type on productivity score. On average in-person and online trainings will lead to different productivity outcomes.  

|Main Effect of Work Hours|<div style="width: 700px">|
|:---|:---|
|Sum of Squares |$4730.625$|
|Degrees of Freedom |$1.0$|
|F-value|$347.49$|
|p-value|$4.44\times10^{-20}$|  

**Interpretation**  
There is an extremely significant main effect of work hours on productivity score. On average 8 vs. 6 work hours will lead to different productivity outcomes.  

|Interaction Effect of<br>Training Type x Work Hours|<div style="width: 700px">|
|:---|:---|
|Sum of Squares |$18.225$|
|Degrees of Freedom |$1.0$|
|F-value|$1.34$|
|p-value|$2.55\times10^{-01}$|  

**Interpretation**  
There is a weak, significant interaction effect. This means that the combination with which training type and work hours are applied matters a small amount to the overall productivity.


|Residual (Error Term)|<div style="width: 700px">|
|:---|:---|
|Sum of Squares |$490.100$|
|Degrees of Freedom |$36.0$|  

**Interpretation**  
This represents the unexplained variability in productivity score after accounting for training type, work hours, and their interaction. Since the residual sum of squares is high relative to the interaction effect this indicates that the interaction effect is weak or non-existent. 

## Story 4 -- Evaluating the Performance of Students
You are conducting an experiment to examine the effect of two factors on the performance of students: teaching method and study time. The two teaching methods are Method A and Method B, and the study times are 2 hours and 4 hours. The goal is to analyze how the teaching method and study time (both separately and in combination) influence the students’ exam scores. You will perform a Two-Way ANOVA to analyze the data. The exam scores for each combination of teaching method and study time are given below:
- Method A, 2 hours: 65, 68, 72, 74, 69, 71, 73, 70, 72, 74, 67, 68, 71, 75, 76, 73, 72, 70, 69, 74
- Method A, 4 hours: 85, 88, 92, 90, 87, 89, 91, 92, 93, 90, 88, 85, 91, 92, 90, 89, 86, 91, 92, 88
- Method B, 2 hours: 60, 63, 64, 62, 61, 60, 59, 62, 65, 64, 66, 63, 61, 60, 62, 61, 63, 65, 64, 62
- Method B, 4 hours: 80, 83, 85, 84, 82, 86, 88, 90, 87, 85, 83, 84, 86, 89, 88, 90, 85, 87, 86, 89

Please answer the following questions:
1. State the null and alternative hypotheses,
2. Check the assumptions before conducting the Two-Way ANOVA,
3. Use Python to solve the question
4. Interpret the results and make conclusions,
5. Visualize the ANOVA test with Python

**Hypotheses**
- $H_0$: The **Null Hypothesis** is there is no significant effect of teaching method or study time on student performance.
  - $H_0^{Method}$ There is no effect of teaching method on student performance.
  - $H_0^{Time}$ There is no effect of study time on student performance.
  - $H_0^{Interaction}$ There is no interaction effect between teaching method and study time.
- $H_1$: The **Alternative Hypothesis** is at least one factor or their interaction has a significant effect on student performance.
  - $H_1^{Method}$ There is a significant effect of teaching method on student performance.
  - $H_1^{Time}$ There is a significant effect of study time on student performance.
  - $H_1^{Interaction}$ There is a significant interaction effect between teaching method and study time.

```python
group_A_2 = pd.DataFrame({'Teaching Method':'Method A','Study Time':'2','Student Performance':[65, 68, 72, 74, 69, 71, 73, 70, 72, 74, 67, 68, 71, 75, 76, 73, 72, 70, 69, 74]})
group_A_4 = pd.DataFrame({'Teaching Method':'Method A','Study Time':'4','Student Performance':[85, 88, 92, 90, 87, 89, 91, 92, 93, 90, 88, 85, 91, 92, 90, 89, 86, 91, 92, 88]})
group_B_2 = pd.DataFrame({'Teaching Method':'Method B','Study Time':'2','Student Performance':[60, 63, 64, 62, 61, 60, 59, 62, 65, 64, 66, 63, 61, 60, 62, 61, 63, 65, 64, 62]})
group_B_4 = pd.DataFrame({'Teaching Method':'Method B','Study Time':'4','Student Performance':[80, 83, 85, 84, 82, 86, 88, 90, 87, 85, 83, 84, 86, 89, 88, 90, 85, 87, 86, 89]})
df = pd.concat([group_A_2, group_A_4, group_B_2, group_B_4])


plt.figure(figsize=(12, 5))
sns.boxplot(x='Teaching Method',y='Student Performance',hue='Study Time',data=df)
plt.title('Student Performance by Teaching Method and Study Time')
plt.show()
```

```python
df = df.rename(columns=lambda s: s.replace("_", " "))

sns.pointplot(x='Teaching Method', y='Student Performance', hue='Study Time', data=df, errorbar=('ci', False))
plt.title('Interaction of Teaching Method and Study Time on Student Performance')
plt.show()
```

```python
df = df.rename(columns=lambda s: s.replace(" ", "_"))

model = ols('Student_Performance ~ C(Teaching_Method) + C(Study_Time) + C(Teaching_Method):C(Study_Time)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
anova_table
```
