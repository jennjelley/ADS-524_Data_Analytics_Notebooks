```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from collections import Counter
from scipy.stats import poisson
```

### Story 1: Heights of Students
The heights of students in a high school are known to follow a normal distribution with a mean height
of 170 cm and a standard deviation of 10 cm.

##### **Generate a sample of 1000 student heights following this distribution**

```python
mean = 170
std_dev = 10
num_samples = 1000

samples = np.random.normal(mean, std_dev, num_samples)
```

##### **Calculate the mean and variance of the generated sample**

```python
mean = np.mean(samples)
variance = np.var(samples)

print(f'\n Mean: {mean}' +
      f'\n Variance: {variance}')
```

##### **Interpret the results**

```python
plt.figure(figsize=(7, 5))
sns.histplot(samples, bins=50, kde=True, color='skyblue')
plt.title('Height Distribution of Students')
plt.xlabel('Height (cm)')
plt.ylabel('Probability Density')
plt.show()
```

### Story 3: Call Center Inbound Calls
A call center receives an average of 3 inbound calls per 10 minutes.

##### **Generate 500 samples from this Poisson distribution**

```python
lambda_value = 3
poisson_samples = np.random.poisson(lam=lambda_value, size=500)
```

##### **Calculate the mean and variance**

```python
mean = np.mean(poisson_samples)
variance = np.var(poisson_samples)

print(f'\n Mean: {mean}' +
      f'\n Variance: {variance}')
```

##### **Interpret their significance for the call center’s operations**

```python
plt.hist(poisson_samples, bins=10, edgecolor='black', alpha=0.7, density=True, label='Sample Data')

x = np.arange(0, max(poisson_samples) + 1)
pmf_values = poisson.pmf(x, mu=lambda_value)

plt.plot(x, pmf_values, marker='o', color='red', label='Theoretical PMF')

plt.title('Distribution of Call Center Inbound Calls')
plt.xlabel('Number of Calls per 10 Minutes')
plt.ylabel('Density')
plt.legend()
plt.show()
```
