## Story 1 -- Student Test Scores
A group of high school students took a math exam, and the teacher wants to analyze the performance to identify areas where students struggle.

The scores of the students are:
- 78, 85, 92, 56, 69, 73, 88, 91, 45, 82, 67, 72, 93, 54, 77, 88, 60, 95, 81, 79, 90, 67, 65, 77, 71, 81, 85, 88, 79, 66, 72, 89, 64, 82, 79, 94, 70, 84, 66, 73

Create a histogram of scores, identify outliers, and calculate mean, median, mode, range, variance, and standard deviation.

```python
import matplotlib.pyplot as plt

student_scores = [
    78, 85, 92, 56, 69, 73, 88, 91, 45, 82, 67, 72, 93, 54, 77, 88, 60, 95, 81, 79, 90, 67, 65, 77, 71, 81, 85, 88, 79, 66, 72, 89, 64, 82, 79, 94, 70, 84, 66, 73]

plt.hist(student_scores, bins=7, edgecolor='black')
plt.title('Histogram of Score Distribution')
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.boxplot(student_scores)
plt.title('Box/Scatter Plot of Score Distribution')
plt.ylabel('Scores')

plt.scatter([1]*len(student_scores), student_scores, color='red', zorder=5)
plt.show()
```

### *Conclusion*
Based on the plot above it is apperant that $45$ is an outlier.

```python
import numpy as np
from collections import Counter

mea = np.mean(student_scores)
med = np.median(student_scores)
mod = Counter(student_scores).most_common(1)[0][0]
ran = np.max(student_scores) - np.min(student_scores)
var = np.var(student_scores)
std = np.std(student_scores)

print(f'\n Mean: {mea} \n Median: {med} \n Mode: {mod} \n' +
      f'\n Range: {ran} \n Variance: {var} \n Standard Deviation: {std} \n')
```

## Story 2 -- Employee Performance Ratings
A company conducts annual performance reviews, and HR collects ratings to determine the overall employee satisfaction and identify top performers. 

The employee performance ratings (1-5 scale) are:
- 3, 4, 5, 2, 3, 5, 4, 2, 3, 4, 5, 1, 3, 5, 4, 2, 4, 3, 3, 5, 4, 3, 2, 4, 5

Create a pie chart of ratings distribution and calculate the measures of center and spread.

```python
import pandas as pd

ratings = [3, 4, 5, 2, 3, 5, 4, 2, 3, 4, 5, 1, 3, 5, 4, 2, 4, 3, 3, 5, 4, 3, 2, 4, 5]

bins = [1, 2, 3, 4, 5, 6]
labels = ['1', '2', '3', '4', '5']

rating_groups = pd.cut(ratings, bins=bins, labels=labels, right=False)
freq_table = rating_groups.value_counts().sort_index()

plt.pie(freq_table, autopct='%1.1f%%', labels=labels)
plt.title('Employee Performance Ratings')
plt.show()
```

```python
mea = np.mean(ratings)
med = np.median(ratings)
mod = Counter(ratings).most_common(1)[0][0]
ran = np.max(ratings) - np.min(ratings)
var = np.var(ratings)
std = np.std(ratings)

print(f'\n Measures of Center' +
      f'\n Mean: {mea} \n Median: {med} \n Mode: {mod} \n' +
      f'\n Measures of Spread' +
      f'\n Range: {ran} \n Variance: {var} \n Standard Deviation: {std} \n')
```

## Story 5 -- Online Course Enrollments
An online learning platform tracks enrollments in its courses to evaluate the demand for different subjects.

||||||||||||
| - | - | - | - | - | - | - | - | - | - | - |
| Course | Python Programming | Data Science | Web Development | Machine Learning | Digital Marketing | Graphic Design | Cybersecurity | AI Basics | Business Analytics | JavaScript Essentials |
| Enrollments | 320 | 450 | 380 | 500 | 310 | 280 | 360 | 400 | 390 | 410 |
| Course | Cloud Computing | Excel for Beginners | Project Management | Python for Data Analysis | UIUX Design | Finance Fundamentals | SQL Basics | Programming | Public Speaking | SEO Optimization |
| Enrollments | 300 | 370 | 420 | 460 | 350 | 330 | 340 | 310 | 295 | 375 |

Use Python to create a bar graph to visualize enrollments by course, identify any trends
in course popularity, and calculate the mean, median, and mode of enrollments.

```python
courses = [
    "Python Programming", "Data Science", "Web Development", "Machine Learning",
    "Digital Marketing", "Graphic Design", "Cybersecurity", "AI Basics", 
    "Business Analytics", "JavaScript Essentials", "Cloud Computing", 
    "Excel for Beginners", "Project Management", "Python for Data Analysis", 
    "UIUX Design", "Finance Fundamentals", "SQL Basics", "Programming", 
    "Public Speaking", "SEO Optimization"
]
enrollments = [
    320, 450, 380, 500, 310, 280, 360, 400, 390, 410,
    300, 370, 420, 460, 350, 330, 340, 310, 295, 375
]

df = pd.DataFrame({'Courses': courses, 'Enrollments': enrollments})
df = df.sort_values('Enrollments')

courses = df['Courses'].to_list()
enrollments = df['Enrollments'].to_list()

plt.bar(courses, enrollments, color='skyblue')
plt.title('Course Enrollments')
plt.xlabel('Course')
plt.xticks(rotation=90)
plt.ylabel('Enrollments')
plt.show()
```

### *Conclusions*
Based on the above graph the 3 most popular classes are data science classes. It appears based on the course titles that introductory and less specialized classes are amoung the least popular.

```python
mea = np.mean(enrollments)
med = np.median(enrollments)
mod = Counter(enrollments).most_common(1)[0][0]

print(f'\n Mean: {mea} \n Median: {med} \n Mode: {mod} \n')
```
